# 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python
# Check for missing values in the 'Rating' column
missing_rating = gpa_cleaned['Rating'].isnull()

# Count the number of missing values
num_missing = missing_rating.sum()

# Print the number of missing values
print("Number of missing values in 'Rating' column:", num_missing)
```

```python
# Drop rows with missing values in the 'Rating' column
gpa_no_missing_rating = gpa_cleaned.dropna(subset=['Rating'])

# Create a new column 'Rating_category'
gpa_no_missing_rating['Rating_category'] = gpa_no_missing_rating['Rating'].apply(lambda x: 'high' if x > 3.5 else 'low')

# Print the shape of the dataset after removing missing values
print("Shape of the dataset without missing values:", gpa_no_missing_rating.shape)
```

It successfully removed the rows with missing values in the 'Rating' column and created a new 'Rating_category' column based on the rating values. The dataset now has a shape of (8892, 14), indicating that you have 8892 rows and 14 columns after these operations.

# 8. Convert the column "Reviews'' to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation) 

```python
# Convert 'Reviews' column to numeric
gpa_cleaned['Reviews'] = pd.to_numeric(gpa_cleaned['Reviews'], errors='coerce')

# Check for outliers using box plot
plt.figure(figsize=(6,4))
sns.boxplot(x=gpa_cleaned['Reviews'])
plt.title('Box Plot of Reviews')
plt.xlabel('Reviews');
```

```python
# Apply log transformation to handle outliers
gpa_cleaned['Reviews_log'] = np.log1p(gpa_cleaned['Reviews'])

# Check for outliers in the transformed column
plt.figure(figsize=(6,4))
sns.boxplot(x=gpa_cleaned['Reviews_log'])
plt.title('Box Plot of Reviews (Log Transformed)')
plt.xlabel('Reviews (Log Transformed)');
```

# 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')

```python
# Replace 'M' with 1 million and 'K' with 1 thousand, and convert to numeric
gpa_no_missing_rating['Size'] = gpa_no_missing_rating['Size'].replace(['M', 'K'], ['', ''], regex=True)
gpa_no_missing_rating['Size'] = pd.to_numeric(gpa_no_missing_rating['Size'], errors='coerce')

# Drop entries where size is 'Varies with device'
gpa_cleaned_size = gpa_no_missing_rating[gpa_no_missing_rating['Size'] != 'Varies with device']

# Print the shape of the dataset after size treatment
print("Shape of the dataset after size treatment:", gpa_cleaned_size.shape)
```
